# ***Exploring the Central Limit Theorem: A Monte Carlo Simulation Approach Across Various Distributions***
This notebook explores the Central Limit Theorem (CLT) using Monte Carlo simulations across different distributions: Bernoulli, Poisson, Beta, and the Binomial proportions. The aim is to observe how the CLT manifests as sample size increases and analyze the performance of confidence intervals.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

**Problem 1: Verifying CLT for Bernoulli Trials**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

def monte_carlo_clt_bernoulli(R, n_values, p_values):
    for p in p_values:
        for n in n_values:
            success_counts = np.random.binomial(n, p, R)
            p_hat = success_counts / n

            mean_p_hat = np.mean(p_hat)
            sd_p_hat = np.sqrt(p * (1 - p) / n)

            sns.histplot(p_hat, kde=True, stat='density', bins=30, label=f'n={n}, p={p}', color='blue')
            x_vals = np.linspace(min(p_hat), max(p_hat), 100)
            plt.plot(x_vals, norm.pdf(x_vals, mean_p_hat, sd_p_hat), label='Normal Fit', color='red')
            plt.title(f'Histogram of Proportion of Successes for n={n}, p={p}')
            plt.xlabel('Proportion of Successes')
            plt.ylabel('Density')
            plt.legend()
            plt.show()

            z_values = [1, 1.64, 1.96, 2.57]
            for z in z_values:
                interval_low = p - z * sd_p_hat
                interval_high = p + z * sd_p_hat
                proportion_in_interval = np.mean((p_hat >= interval_low) & (p_hat <= interval_high))
                print(f'For p={p}, n={n}, within Â±{z}sd: {proportion_in_interval:.3f}')

R = 10000
n_values = [20, 50, 100, 200, 300, 500]
p_values = [0.1, 0.3, 0.5]

monte_carlo_clt_bernoulli(R, n_values, p_values)
```



*   As sample size increases, the distribution of sample proportions approaches a normal distribution.
*   The empirical proportions within Â±Z standard deviations align with theoretical expectations as n increases.



**Problem 2: CLT for Poisson Distribution**

```python
def monte_carlo_clt_poisson(R, n_values, mu_values):
    for mu in mu_values:
        for n in n_values:
            samples = np.random.poisson(mu, (R, n))
            sample_means = np.mean(samples, axis=1)

            mean_sample_means = np.mean(sample_means)
            sd_sample_means = np.std(sample_means)

            sns.histplot(sample_means, kde=True, stat='density', bins=30, label=f'n={n}, Î¼={mu}', color='green')
            x_vals = np.linspace(min(sample_means), max(sample_means), 100)
            plt.plot(x_vals, norm.pdf(x_vals, mean_sample_means, sd_sample_means), label='Normal Fit', color='red')
            plt.title(f'Histogram of Sample Means for n={n}, Î¼={mu}')
            plt.xlabel('Sample Mean')
            plt.ylabel('Density')
            plt.legend()
            plt.show()

            z_values = [1, 1.64, 1.96, 2.57]
            for z in z_values:
                interval_low = mu - z * (np.std(sample_means) / np.sqrt(n))
                interval_high = mu + z * (np.std(sample_means) / np.sqrt(n))
                proportion_in_interval = np.mean((sample_means >= interval_low) & (sample_means <= interval_high))
                print(f'For Î¼={mu}, n={n}, within Â±{z}sd: {proportion_in_interval:.3f}')

R = 10000
n_values = [20, 50, 200, 300, 500]
mu_values = [5, 20, 30, 50]

monte_carlo_clt_poisson(R, n_values, mu_values)
```



*   For larger sample sizes, the sample means of Poisson-distributed data approximate a normal distribution, as predicted by the CLT.
*   Proportions within the intervals improve with increasing n, validating the CLT for Poisson data.



**Problem 3: Confidence Intervals for Binomial Proportions**

```python
def monte_carlo_confidence_intervals_binomial(R, n_values, p_values):
    for p in p_values:
        for n in n_values:
            successes = np.random.binomial(n, p, R)
            p_hat = successes / n
            lower_bound = p_hat - 1.96 * np.sqrt(p_hat * (1 - p_hat) / n)
            upper_bound = p_hat + 1.96 * np.sqrt(p_hat * (1 - p_hat) / n)
            contained = np.mean((lower_bound <= p) & (upper_bound >= p))
            print(f'For p={p}, n={n}, proportion of intervals containing true p: {contained:.3f}')

R = 10000
n_values = [20, 50, 200, 300, 500]
p_values = [0.01, 0.2, 0.3, 0.4, 0.5]

monte_carlo_confidence_intervals_binomial(R, n_values, p_values)
```

*   Coverage probability approaches 95% as n increases.
*   For small n and extreme values of p, Wald intervals show undercoverage, especially when p is close to 0 or 1.



**Problem 4: CLT for Beta Distribution**

```python
def monte_carlo_clt_beta(R, n_values, alpha_beta_values):
    for alpha, beta in alpha_beta_values:
        for n in n_values:
            samples = np.random.beta(alpha, beta, (R, n))
            sample_means = np.mean(samples, axis=1)

            mean_sample_means = np.mean(sample_means)
            sd_sample_means = np.std(sample_means)

            sns.histplot(sample_means, kde=True, stat='density', bins=30, label=f'n={n}, Î±={alpha}, Î²={beta}', color='purple')
            x_vals = np.linspace(min(sample_means), max(sample_means), 100)
            plt.plot(x_vals, norm.pdf(x_vals, mean_sample_means, sd_sample_means), label='Normal Fit', color='red')
            plt.title(f'Histogram of Sample Means for n={n}, Î±={alpha}, Î²={beta}')
            plt.xlabel('Sample Mean')
            plt.ylabel('Density')
            plt.legend()
            plt.show()

            z_values = [1, 1.64, 1.96, 2.57]
            for z in z_values:
                interval_low = mean_sample_means - z * (sd_sample_means / np.sqrt(n))
                interval_high = mean_sample_means + z * (sd_sample_means / np.sqrt(n))
                proportion_in_interval = np.mean((sample_means >= interval_low) & (sample_means <= interval_high))
                print(f'For Î±={alpha}, Î²={beta}, n={n}, within Â±{z}sd: {proportion_in_interval:.3f}')

R = 10000
n_values = [20, 50, 200, 300, 500]
alpha_beta_values = [(2, 5), (1, 1), (5, 2)]

monte_carlo_clt_beta(R, n_values, alpha_beta_values)
```



*   The sample means of the Beta distribution tend toward normality as n increases.
*   Proportions within the intervals demonstrate convergence to theoretical expectations with increasing n, reinforcing the CLT.
