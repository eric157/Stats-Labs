```python
def process_id(id):
    year = str(id)[:4]
    if(year == None): return None
    if(year == '0'): return None
    if(year == '2024'): return 1
    if(year == '2023'): return 2
    if(year == '2022'): return 3
    if(year == '2021'): return 4
    else: return None

df['YEAR'] = df['SID'].apply(process_id)
df = df.drop(columns=['SID'], axis=1)
df.head()
```

```python
def process_fam(x):
    if(type(x) is int): return None
    if(type(x) is float): return None
    x = x.lower()
    if('joi' in x): return 'JOINT'
    if('NUC'): return "NUCLEAR"
    else : return None

df['Joint Family?'] = df['Joint Family?'].apply(process_fam)
```

```python
def process_family_size(x):
    try:
        return int(x)
    except Exception as e:
        return None

df['Family Size'] = df['Family Size'].apply(process_family_size)
```

```python
df['Family Income'].value_counts()
```

```python
def process_family_income(x):
    try:
        x = int(x)
        if(x == 0): return None
        if(x < 10): return x * 100000
        else: return x
    except Exception as e:
        if(type(x) is not str): return None
        x = x.split(' ')
        for i in x:
            if(i.isdigit()):
                if(int(i) < 10): return int(i) * 100000
                else: return int(i)
        return None

df['Family Income'] = df['Family Income'].apply(process_family_income)
```

```python
def process_jee_rank(x):
    try:
        x = int(x)
        return x
    except Exception as e:
        if(type(x) is not str): return None
        if 'K' in x:
            x = x.replace('K', '')
            x = float(x)
            x = x * 1000
            return x
        return None

df['Rank JEE'] = df['Rank JEE'].apply(process_jee_rank)
```

```python
def process_cgpa(x):
    try:
        x = float(x)
        return x
    except Exception as e:
        return None

df['Latest CGPA'] = df['Latest CGPA'].apply(process_cgpa)
```

```python
def process_rating(x):

    try:
        x = int(x)
        return x
    except Exception as e:
        if(type(x) is not str): return None
        x = x.lower()
        if('satisfactory' in x): return 3
        if('poor' in x): return 4
        if('very' in x): return 2
        if('excellent' in x): return 1
        else: return None

df["TEACHING"] = df["TEACHING"].apply(process_rating)
df["INFRA"] = df["INFRA"].apply(process_rating)
df["SPORTS"] = df["SPORTS"].apply(process_rating)
df["HOSTEL"] = df["HOSTEL"].apply(process_rating)
```

```python
df['How know?'].value_counts()
```

```python
def process_know(x):
    if(type(x) is not str): return None
    x = x.lower()
    if('friend' in x or 'fri' in x): return 'FRIEND'
    if('teacher' in x or 'school' in x or 'coaching' in x): return 'TEACHER'
    if('family' in x or 'parent' in x or 'brother' in x or 'sister' in x): return 'FAMILY'
    else : return 'INTERNET'

df['How know?'] = df['How know?'].apply(process_know)
```

```python
def process_time(x):
    try:
        x = float(x)
        return x
    except Exception as e:
        if(type(x) is not str): return None
        x = x.split(' ')
        for i in x:
            if(i.isdigit()): return float(i)
            if('-' in i):
                i = i.split('-')
                try:
                    return (float(i[0]) + float(i[1])) / 2
                except Exception as e:
                    return None
        return None

df['CELL PHONE'] = df['CELL PHONE'].apply(process_time)
df['STUDY'] = df['STUDY'].apply(process_time)
df['EXTRA'] = df['EXTRA'].apply(process_time)
df['CO'] = df['CO'].apply(process_time)
df['CHATTING'] = df['CHATTING'].apply(process_time)
df['SLEEPING'] = df['SLEEPING'].apply(process_time)
```

```python
df['Higher study?'] = df['Higher study?'].str.lower().str.strip()

mapping = {
    'no': 'No',
    'yes': 'Yes',
    'maybe': 'Maybe',
    'have not thought about it!': 'Not Decided',
    'not yet decided': 'Not Decided'
}

df['Higher study?'] = df['Higher study?'].replace(mapping)
```
